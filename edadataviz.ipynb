### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Clean accidental trailing spaces from column names
df.columns = df.columns.str.strip()

# 2. Dynamically locate the Flight Number column
flight_col = None
for col in df.columns:
    if col.lower() in ['flightnumber', 'flight number', 'flight no.']:
        flight_col = col
        break
if not flight_col:
    flight_col = df.columns[0] # Fallback to the very first column

# 3. Dynamically locate the Launch Site column
site_col = None
for col in df.columns:
    if col.lower() in ['launchsite', 'launch site']:
        site_col = col
        break
if not site_col:
    # Fallback to look for 'site' in column names
    site_col = [c for c in df.columns if 'site' in c.lower()][0]

print(f"Plotting using detected columns -> X-axis: '{flight_col}', Y-axis: '{site_col}'")

# 4. Ensure Flight Number data is numeric so it plots chronologically
df[flight_col] = pd.to_numeric(df[flight_col], errors='coerce')

# 5. Create the Seaborn Catplot
sns.catplot(
    data=df, 
    x=flight_col, 
    y=site_col, 
    hue='Class', 
    aspect=2.5, 
    height=6,
    palette={1: '#2ca02c', 0: '#d62728'}  # Green for success, Red for failure
)

# 6. Customize labels
plt.title('SpaceX Launch Site vs. Flight Number (Grouped by Landing Success)', fontsize=16, pad=15)
plt.xlabel('Flight Number', fontsize=14)
plt.ylabel('Launch Site / Facility', fontsize=14)

# 7. Show the plot
plt.show()
```

Now try to explain the patterns you found in the Flight Number vs. Launch Site scatter point plots.

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# HINT use groupby method on Orbit column and get the mean of Class column
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Clean up any accidental trailing spaces from column names
df.columns = df.columns.str.strip()

# 2. Dynamically locate the Orbit column
orbit_col = None
for col in df.columns:
    if col.lower() in ['orbit', 'orbit type']:
        orbit_col = col
        break

# 3. Calculate the success rate (mean of Class) for each orbit type
# We reset the index to convert the resulting Series back into a clean DataFrame
orbit_success = df.groupby(orbit_col)['Class'].mean().reset_index()

# Convert the rate to a percentage for better readability
orbit_success['Success Rate (%)'] = orbit_success['Class'] * 100

# Sort the values so the chart reads nicely from highest success rate to lowest
orbit_success = orbit_success.sort_values(by='Success Rate (%)', ascending=False)

# 4. Create the Bar Chart
plt.figure(figsize=(12, 6))
sns.barplot(
    data=orbit_success, 
    x=orbit_col, 
    y='Success Rate (%)', 
    palette='viridis' # A clean, modern gradient color palette
)

# 5. Customize the labels and appearance
plt.title('SpaceX Landing Success Rate by Target Orbit Type', fontsize=16, pad=15)
plt.xlabel('Orbit Type', fontsize=14)
plt.ylabel('Success Rate (%)', fontsize=14)
plt.ylim(0, 105) # Add a bit of space above the 100% mark for text labels

# Optional: Add text labels on top of each bar showing the exact percentage
for index, row in orbit_success.iterrows():
    plt.text(
        x=index, 
        y=row['Success Rate (%)'] + 2, 
        s=f"{row['Success Rate (%)']:.1f}%", 
        color='black', 
        ha="center", 
        fontsize=10
    )

# 6. Show the plot
plt.show()
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
    
```

```python
# Plot a line chart with x axis to be the extracted year and y axis to be the success rate
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Clean up any accidental trailing spaces from column names
df.columns = df.columns.str.strip()

# 2. Extract years from the Date column safely using the requested logic
year = []
def Extract_year():
    # Using str(i) handles cases where some dates might read as timestamp objects
    for i in df["Date"]:
        year.append(str(i).split("-")[0])
    return year

# Run the function and assign the result to a new 'Year' column 
# Note: Keeping the original Date column intact prevents overwriting issues if re-run
df['Year'] = Extract_year()

# 3. Calculate the average success rate per year
yearly_trend = df.groupby('Year')['Class'].mean().reset_index()

# Convert the rate to a percentage for intuitive reading
yearly_trend['Success Rate (%)'] = yearly_trend['Class'] * 100

# 4. Plot the Yearly Trend Line Chart
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=yearly_trend, 
    x='Year', 
    y='Success Rate (%)', 
    marker='o',         # Adds explicit data points to the timeline junctions
    linewidth=2.5, 
    color='#1f77b4'
)

# 5. Customize labels and layout
plt.title('SpaceX Launch Success Rate Yearly Trend', fontsize=16, pad=15)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Success Rate (%)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)  # Gridlines help trace precise values over time
plt.ylim(-5, 105)                          # Ensures the scale ranges properly from 0 to 100%

# 6. Show the plot
plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
# HINT: Use get_dummies() function on the categorical columns
import pandas as pd

# 1. Clean column names to make sure there are no trailing spaces causing KeyErrors
df.columns = df.columns.str.strip()

# 2. Select the specific categorical columns mentioned in your task
# Note: Matching common casing variations (e.g., 'Orbit', 'Launch site', 'Landing pad', 'Serial')
categorical_cols = []
for target in ['orbit', 'launch site', 'launchsite', 'landing pad', 'landingpad', 'serial']:
    matched = [col for col in df.columns if col.lower() == target]
    if matched:
        categorical_cols.append(matched[0])

print(f"Categorical columns found for encoding: {categorical_cols}")

# 3. Apply get_dummies to only the selected categorical columns
features_one_hot = pd.get_dummies(df[categorical_cols], dtype=int)

# 4. Drop the original categorical columns from the main dataframe 
# and concatenate the newly generated binary dummy columns
features_final = pd.concat([df.drop(columns=categorical_cols), features_one_hot], axis=1)

# 5. Display the results using the head() method
print(f"\nNew DataFrame Shape: {features_final.shape}")
features_final.head()
```
